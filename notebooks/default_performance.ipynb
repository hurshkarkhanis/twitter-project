```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
%config InlineBackend.figure_format = 'retina'
```

```python
numbers = {
'Random Forest':0.7715877437325905,
'Gradient BoostingClassifier':0.6601671309192201,
'Decision TreeClassifier':0.7270194986072424,
'Logistic Regression':0.8022284122562674,
'K-Neighbors':0.6323119777158774,
}
```

```python
fig, ax = plt.subplots(figsize=(17, 10))
ax.bar(numbers.keys(), numbers.values())
ax.set_ylabel("score", fontsize=20)
ax.set_xlabel("model", fontsize=20)
ax.set_title("Accuracy Score with Default Parameters", fontsize=20)
ax.tick_params(axis='x', labelsize= 15)
ax.tick_params(axis='y', labelsize= 15)
plt.savefig('../graphs/final_default.png')
```
